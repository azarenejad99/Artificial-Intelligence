```python
def RMSE_calc(actual_y,predict_y):
    N = len(actual_y)
    square_sum = 0
    for i in range(N):
        temp = (predict_y[i] - actual_y[i])**2
        square_sum += temp
    square_sum /= N
    return float(square_sum**0.5)
    
    
```

```python
def linear_calc(x,w = 40.16235654 , b=-68.35748451752235  ):
    return x*w + b
```

```python
y = df['SalePrice'].tolist()
```

```python
def make_predict_y_list():
    X = df['OverallQual'].tolist()
    for i in range(len(X)):
        X[i] =  linear_calc(X[i])
    return X
        
```

```python
y_pred = make_predict_y_list()
```

```python
rmse =  RMSE_calc(y,y_pred)
print(rmse)
```
