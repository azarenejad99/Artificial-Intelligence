حال کافی است که در قسمت نهایی بخش ترین خود به محاسبه احتمال وجود کلمه به شرط انکه شاعرش حافظ یا سعدی باشد می پردازیم .در حقیقت با استفاده از فرمول زیر احتمال را محاسبه می کنیم.

p ( w | hafez ) =  تعداد دفعاتی که کلمه مورد نظر در حافظ آمده +۱/  تعداد کماتی که در حافظ آمده + تعداد کلماتی که در کل آمده 

p ( w | saadi ) =  تعداد دفعاتی که کلمه مورد نظر در سعدی آمده +۱/  تعداد کماتی که در سعدی آمده + تعداد کلماتی که در کل آمده 

حال بر اساس فرمول های فوق به محاسبه احتمال برای هر کلمه به شرط هر حافظ می پردازیم
در بالا عبارت لاپلاس به کار رفت که حالا در مورد آن بحث می کنیم.
وقتی کلمه کشتی تو بیت های ترین شده سعدی به کار نرفته باشه ولی تو بیت های تستش باشه اینو که ضرب کنبم احتمال کل صفر میشه بدون اینکه بقیه رو در نظر بگیره ولی با اضافه کردن یک حتمال دیگر صفر نمیشود.

```python
def calc_prob_conditional_hafez_laplace(word,hafez_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_hafez_not_unique , total_vocab_size_unique_word):
    num_word_in_hafez = 0
    if word in hafez_vocab_in_train_data:
        num_word_in_hafez = hafez_vocab_in_train_data[word]
    x =  num_word_in_hafez + 1
    y = float( num_total_word_hafez_not_unique + total_vocab_size_unique_word)
    return x/y
    
    


def calc_prob_conditional_hafez_simple(word,hafez_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_hafez_not_unique , total_vocab_size_unique_word):
    num_word_in_hafez = 0
    if word in hafez_vocab_in_train_data:
        num_word_in_hafez = hafez_vocab_in_train_data[word]
    x = num_word_in_hafez
    y = float(num_total_word_hafez_not_unique )
    return x/y
    

def calc_prob_conditional_saadi_laplace(word,saadi_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_saadi_not_unique , total_vocab_size_unique_word  ):
    num_word_in_saadi = 0
    if word in saadi_vocab_in_train_data:
        num_word_in_saadi = saadi_vocab_in_train_data[word]
    x =  num_word_in_saadi + 1
    y = float( num_total_word_saadi_not_unique + total_vocab_size_unique_word)
    return x/y
    

def calc_prob_conditional_saadi_simple(word,saadi_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_saadi_not_unique , total_vocab_size_unique_word):
    num_word_in_saadi = 0
    if word in saadi_vocab_in_train_data:
        num_word_in_saadi = saadi_vocab_in_train_data[word]
    x = num_word_in_saadi 
    y = float(num_total_word_saadi_not_unique )
    return x/y
    
```

حال در این قسمت با توجه به دیتا ترین کافی است که از ۲۰ درصد باقی مانده برای تست عملکرد استفاده کنیم .در اینجا سه مقدار را براساس فرمول های زیر بدست خواهمیم آورد

در اینجا کافی است در هر اندیس دیتا تست بیاییم هر شعر را بررسی کنیم و  دو مقدار زیر را هم برای حافظ و هم برای سعدی حساب کنیم  و هر کدام که بیشتر شد در حقیقت مدل ما آن را پیش بینی کرده است حال با نتیجه جواب مقاسه می کنیم و مقادیر را آپدیت می کنیم

ابتدا تست را با حالت لاپلاس بررسی می کنیم

```python
def test_laplace(statuses,
             poems,
             probability_hafez,
             probability_saadi,
            total_vocab_in_train_data,
            total_vocab_size_unique_word,
            num_total_word_vocab_not_unique,
            hafez_vocab_in_train_data,
            saadi_vocab_in_train_data,
            num_total_word_hafez_unique,
            num_total_word_saadi_unique,
            num_total_word_hafez_not_unique,
            num_total_word_saadi_not_unique):
    
    hafez_poems_in_test_data = []
    saadi_poems_in_test_data = []

    for index in test_index :
        if statuses[index] == 'hafez' :
            hafez_poems_in_test_data.append(poems[index])
        else :
            saadi_poems_in_test_data.append(poems[index])
            
    hafez_poems_num_in_test_data = len(hafez_poems_in_test_data)
    saadi_poems_num_in_test_data = len(saadi_poems_in_test_data)

    print("hafez poems count in 20% train data:",hafez_poems_num_in_test_data)
    print("saadi poems count in 20% train data:",saadi_poems_num_in_test_data)
    total_vocab_in_test_data = {}

    for index in test_index:
        poem = poems[index]
        words = poem.split(' ')
        for word in words:
            if word not in total_vocab_in_test_data:
                total_vocab_in_test_data[word] = 0
            else:
                total_vocab_in_test_data[word]+=1
            
    total_vocab_size_unique_word_in_test = len(total_vocab_in_test_data)

    num_total_word_vocab_not_unique_in_test = 0
    for key in total_vocab_in_test_data :
        num_total_word_vocab_not_unique_in_test += total_vocab_in_test_data[key]
    
    print("num_total_word_vocab_not_unique in 20% test",num_total_word_vocab_not_unique_in_test)

    print("total_vocab_size_unique_word in 20 % test",total_vocab_size_unique_word_in_test)
    print("num poem in train data in 20% test",len(test_index))
    
    hafez_vocab_in_test_data = {}
    saadi_vocab_in_test_data = {}



    for poem in hafez_poems_in_test_data:
        words = poem.split(' ')
        for word in words:
            if word not in hafez_vocab_in_test_data:
                hafez_vocab_in_test_data[word] = 0
            else:
                hafez_vocab_in_test_data[word]+=1


    for poem in saadi_poems_in_test_data :
        words = poem.split(' ')
        for word in words:
            if word not in saadi_vocab_in_test_data:
                saadi_vocab_in_test_data[word] = 0
            else:
                saadi_vocab_in_test_data[word]+=1    

                
                
    num_total_word_hafez_unique_in_test = len(hafez_vocab_in_test_data) 
    num_total_word_hafez_not_unique_in_test = 0
    for key in hafez_vocab_in_test_data :
        num_total_word_hafez_not_unique_in_test += hafez_vocab_in_test_data[key]
    
    num_total_word_saadi_unique_in_test = len(saadi_vocab_in_test_data) 

    num_total_word_saadi_not_unique_in_test = 0
    for key in saadi_vocab_in_test_data :
        num_total_word_saadi_not_unique_in_test += saadi_vocab_in_test_data[key]

    print("num_total_word_hafez_not_unique in 20% test:",num_total_word_hafez_not_unique_in_test)
    print("num_total_word_hafez_unique in 20% test: ",num_total_word_hafez_unique_in_test)
    print("num_total_word_saadi_not_unique in 20% test:",num_total_word_saadi_not_unique_in_test  )
    print("num_total_word_saadi_unique in 20 % test",num_total_word_saadi_unique_in_test)
            
        
    correct_detected_hafezes = 0
    all_hafezes = len(hafez_poems_in_test_data)
    detected_hafezes = 0
    correct_detected = 0
    total = len(test_index)
    
    for index in test_index:
        poem = poems[index]
#         print(poem)
#         print(statuses[index])
        words = poem.split(' ')
        p_hafez = probability_hafez
        p_saadi = probability_saadi
        for word in words:
            p_hafez *= calc_prob_conditional_hafez_laplace(word,hafez_vocab_in_train_data,
                                                           total_vocab_in_train_data,
                                                            num_total_word_hafez_not_unique ,
                                                            total_vocab_size_unique_word)
            p_saadi *= calc_prob_conditional_saadi_laplace(word,saadi_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_saadi_not_unique , total_vocab_size_unique_word  )

            
        if p_hafez > p_saadi:
#             print("detected_hafezess")
            detected_hafezes += 1
            if statuses[index] == 'hafez':
#                 print("true")
                correct_detected_hafezes += 1
                correct_detected += 1
            else:
                pass
#                 print("false")
          
        ##model decide saadi
        elif p_hafez < p_saadi:
#             print("detected_saadi")
            if statuses[index] =='saadi':
#                 print("true")
                correct_detected+=1
            else:
                pass
#                 print("false")
                

    recall = correct_detected_hafezes/float(all_hafezes)
    precision =  correct_detected_hafezes/ float(detected_hafezes)  
    accuracy = correct_detected / float(total)
            
            
    print("Recall in laplace:",recall)
    print("Precision in laplace:",precision)
    print("Accuracy in laplace:",accuracy)

test_laplace(statuses,
             poems,
             probability_hafez,
             probability_saadi,
            total_vocab_in_train_data,
            total_vocab_size_unique_word,
            num_total_word_vocab_not_unique,
            hafez_vocab_in_train_data,
            saadi_vocab_in_train_data,
            num_total_word_hafez_unique,
            num_total_word_saadi_unique,
            num_total_word_hafez_not_unique,
            num_total_word_saadi_not_unique)
        
        
```

```python
def make_evalute_file( evalute_filename ,
                    probability_hafez,
                    probability_saadi,
                    total_vocab_in_train_data,
                    total_vocab_size_unique_word,
                    num_total_word_vocab_not_unique,
                    hafez_vocab_in_train_data,
                    saadi_vocab_in_train_data,
                    num_total_word_hafez_unique,
                    num_total_word_saadi_unique,
                    num_total_word_hafez_not_unique,
                    num_total_word_saadi_not_unique):
    
    Data = pd.read_csv(evalute_filename)
    index = Data['id']
    poems = Data['text']
    statuses = []
    for poem in poems:
        words = poem.split(' ')
        p_hafez = probability_hafez
        p_saadi = probability_saadi
        for word in words:
            p_hafez *= calc_prob_conditional_hafez_laplace(word,hafez_vocab_in_train_data,
                                                           total_vocab_in_train_data,
                                                            num_total_word_hafez_not_unique ,
                                                            total_vocab_size_unique_word)
            p_saadi *= calc_prob_conditional_saadi_laplace(word,saadi_vocab_in_train_data,total_vocab_in_train_data,
                                        num_total_word_saadi_not_unique , total_vocab_size_unique_word  )
            
        if p_hafez > p_saadi:
            statuses.append('hafez')
        elif p_hafez < p_saadi:
            statuses.append('saadi')
    Data.insert(2,'label',statuses)
    print(Data.head(5))
    Data.to_csv("output.csv")
    
```
